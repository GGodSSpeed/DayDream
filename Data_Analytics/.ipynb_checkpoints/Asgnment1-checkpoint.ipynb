6. Make a countplot of Movies directed by top 5 directors.

```python
top = df.groupby(['Director'])['IMDB_Rating'].mean().nlargest(5)
sb.barplot(y=top.index, x=top.values).set(xlabel="IMDB")
```

7. Make a plot to analyze distribution of IMDB rating?

```python
plt.figure(figsize=(10,5))
# IMDB_plot = df['IMDB_Rating'].plot().set(ylabel="IMDB")
sb.barplot(y=df['IMDB_Rating'], x=df['Meta_score'])
```

8. What type of Genre has the highest IMDB rating?

```python
plt.figure(figsize=(20,10))
imdb = df.set_index('IMDB_Rating')
imdb_plot = pd.DataFrame(imdb['Genre'].str.split(", ").explode())
imdb_plot.reset_index(inplace=True)
final_imdb = imdb_plot.groupby('Genre').count()
print(final_imdb.nlargest(1, columns='IMDB_Rating'))
sb.barplot(y = final_imdb['IMDB_Rating'], x=final_imdb.index)
```

9. Name the top 5 movies that have earned highest Meta score. 

```python
plt.figure(figsize=(15,10))
metascore_top = df.sort_values(['Meta_score', 'IMDB_Rating'], ascending=False)
meta_plot = sb.barplot(x = metascore_top.head().index, y = metascore_top.head()['Meta_score'])
```
