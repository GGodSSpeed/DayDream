```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pict = plt.imread("Milky_way.jpg")
plt.imshow(pict)
```

```python
grayscale_img = pict.mean(axis = 2)     #Pixel mean along the Z-axis
imgplot = plt.imshow(grayscale_img, cmap = "gray")
```

### Another way to obtain Grayscale image

```python
rgb_weights = [0.2989, 0.5870, 0.1140]      #Mean weights of the RGB channels
grayscale_img = np.dot(pict, rgb_weights)   #Dot product
imgplot = plt.imshow(grayscale_img, cmap="gray")
```

```python
final_image = np.where(grayscale_img < 150, 0, 255) #SComparison and assignment
plt.imsave("Grayscaled.jpg", final_image)
plt.imshow(final_image, cmap='gray') 
```

```python
pixel_mean = np.mean(final_image)
print(pixel_mean)
```
