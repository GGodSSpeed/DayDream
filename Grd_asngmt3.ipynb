```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pict = plt.imread('Milky_way.jpg')  #Reading Images..
plt.figure(figsize=(10,10)) #Sets the size of the figure displayed..
plt.imshow(pict)
```

```python
experimental_pict = pict.copy()
experimental_pict[experimental_pict[:,:,0]<50] = 0
final_image = experimental_pict
plt.imsave("grd_img3.jpg", final_image)
plt.imshow(final_image)
```

```python
red_channel = final_image[:,:,0]
null_pixel_count = np.count_nonzero(red_channel==0) #Counts non-zero values for iterable element
print(null_pixel_count)
```
